1. What is Python, and why is it popular?
  - Python is a widely-used programming language known for its simplicity and versatility. It's popular because it's easy to read and write, has a large community, and supports many applications like web development, data analysis, AI, and automation.

2. What is an interpreter in Python?
   - An interpreter is a program that reads Python code and executes it line by line, converting it into a form the computer understands.

3. What are pre-defined keywords in Python?
   - Keywords are special words in Python (like if, for, return) which have special meanings and cannot be used for any other purpose.

4. Can keywords be used as variable names?
   - No, keywords cannot be used as variable names because they are reserved by Python for specific functions.

5. What is mutability in Python?
   - Mutability refers to the ability of an object to change its value after it's created. Mutable objects can be changed (like lists), while immutable ones cannot be changed (like tuples and strings).

6. Why are lists mutable, but tuples are immutable?   
   - Lists are designed for cases where data changes, so their elements can be modified. Tuples, however, are made to store data that should remain constant, so their elements can't be modified after creation.

7. What is the difference between “==” and “is” operators in Python?
   - == checks if the values are equal, while is checks if two variables point to the same object in memory.

8. What are logical operators in Python?
   - Logical operators are used to combine multiple conditions. The main ones are and, or, and not.

   and: True if both statements are true

   or: True if at least one statement is true

   not: Reverses the truth value

9. What is type casting in Python?
   - Type casting means converting a value from one data type to another, like turning a string "12" into the integer 12.

10. What is the difference between implicit and explicit type casting?

  - Implicit type casting is done automatically by Python when needed.

  - Explicit type casting is when the programmer manually converts data types using functions like int(), str(), or float().

11. What is the purpose of conditional statements in Python?
   - Conditional statements help a program make decisions by executing different blocks of code based on certain conditions.

12. How does the elif statement work?
   - elif (short for “else if”) allows you to check multiple conditions one after the other. It's used after an if and before an else.

13. What is the difference between for and while loops?

   - A for loop is used when you know how many times you want to loop (like looping over a fixed list).

   A while loop is used when you don't know in advance how many times you'll need to loop; it repeats as long as a condition is true.

14. Describe a scenario where a while loop is more suitable than a for loop.
   - If you’re asking the user to keep entering data until they enter 'quit', a while loop is best because you don't know how many times the user will input data.
