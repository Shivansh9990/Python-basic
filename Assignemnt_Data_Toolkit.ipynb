What is NumPy, and why is it widely used in Python?
NumPy is a foundational Python library providing support for large, multi-dimensional arrays and matrices plus a large collection of mathematical functions to operate on them efficiently. It is widely used for its speed, capability for vectorized operations, and ease of integration with other scientific computing libraries.

How does broadcasting work in NumPy?
Broadcasting allows NumPy to perform arithmetic operations on arrays of different shapes by automatically expanding the smaller array to match the shape of the larger array without copying data, enabling efficient and concise computations.
import numpy as np
a = np.array([1,2,3])
b = 2
print(a + b)  # [3 4 5]

What is a Pandas DataFrame?
A DataFrame is a two-dimensional, size-mutable, heterogeneous data structure with labeled axes (rows and columns) in Pandas, commonly used for data cleaning, manipulation, and analysis.

Explain the use of the groupby() method in Pandas.
groupby() splits data into groups based on some criteria, allowing for aggregation, transformation, or filtration of data within those groups, essential for data summarization.
df.groupby("department")["salary"].mean()

Why is Seaborn preferred for statistical visualizations?
Seaborn simplifies the creation of aesthetically pleasing statistical graphics with built-in themes and color palettes, making data insights clearer and more visually effective.

Differences between NumPy arrays and Python lists.
NumPy arrays are homogeneous (same data type), fixed size, and support fast, vectorized operations. Python lists are heterogeneous, dynamic in size, and slower for numerical tasks due to lack of optimization.

What is a heatmap, and when should it be used?
A heatmap visualizes data where individual values are represented as colors, often used for correlation matrices or frequency tables where patterns are discerned visually.

What does “vectorized operation” mean in NumPy?
It refers to operations applied element-wise on entire arrays without explicit loops, leveraging low-level optimizations for speed.
a = np.array([1,2,3])
b = np.array([4,5,6])
print(a+b)  # [5 7 9]

How does Matplotlib differ from Plotly?
Matplotlib creates static plots mainly for offline use, while Plotly generates interactive, web-embeddable visuals with zoom, hover, and dynamic updates.

What is the significance of hierarchical indexing in Pandas?
Enables multiple index levels on an axis, facilitating complex data representation and easier data slicing or selection.

What is the role of Seaborn’s pairplot() function?
It generates a matrix of scatterplots and histograms to visualize pairwise relationships between several variables.

What is the purpose of the describe() function in Pandas?
Provides descriptive statistics (count, mean, std, min, max, quartiles) for numerical columns to summarize data quickly.

Why is handling missing data important in Pandas?
Unaddressed missing data can bias analyses or cause computational errors; managing them ensures cleaner data and reliable results.

What are the benefits of using Plotly for data visualization?
Interactive, user-friendly graphics with zooming, hovering, exporting capabilities for flexible and rich data exploration.

How does NumPy handle multidimensional arrays?
Through ndarrays supporting efficient storage, slicing, reshaping, and broadcasted operations on multi-dimensional data.
arr = np.array([[1,2],[3,4]])
print(arr.shape)  # (2,2)

What is the role of Bokeh in data visualization?
Produces interactive web visualizations optimized for streaming data and dashboards.

Explain the difference between apply() and map() in Pandas.
apply() executes a function along DataFrame rows or columns; map() substitutes values in a Series based on a correspondence (dictionary or function).

What are some advanced features of NumPy?
Broadcasting, masked arrays, structured arrays, FFT, linear algebra routines, random number generation, and C/C++ integration.

How does Pandas simplify time series analysis?
With datetime indexing, resampling, window functions, and frequency conversion tools tailored for time series data.

What is the role of a pivot table in Pandas?
Allows data aggregation and reshaping by summarizing data across multi-dimensional categorical variables.

Why is NumPy’s array slicing faster than Python’s list slicing?
Because NumPy slicing operates on contiguous memory blocks at the C level without copying, whereas list slicing copies references at the Python level.

What are some common use cases for Seaborn?
Distribution plots, categorical plots, regression visualizations, heatmaps, and multivariate matrix plots.
