1. What are data structures, and why are they important?
Data structures are specialized formats for organizing, processing, and storing data. They help manage large amounts of data efficiently for various tasks such as searching, sorting, and modifying data.

2. Difference between mutable and immutable data types (with examples)
Mutable: Can be changed after creation (e.g., lists – my_list = ).

Immutable: Cannot be changed after creation (e.g., strings – my_str = "abc", tuples – my_tuple = (1, 2, 3)).

3. Main differences between lists and tuples in Python
lists are Mutable, tuples are Immutable.

4. How dictionaries store data
Dictionaries store data as key-value pairs, using hash functions to associate each key with a value for fast retrieval.

5. Why use a set instead of a list in Python?
Sets automatically remove duplicates and provide fast membership testing, ideal when uniqueness is required.

6. What is a string in Python, and how is it different from a list?
A string is a sequence of characters (immutable).

A list is a sequence of items (mutable, can be changed and hold mixed types).

7. How do tuples ensure data integrity in Python?
Tuples are immutable, meaning their values cannot be changed, thus protecting data from accidental modification.

8. What is a hash table, and how does it relate to dictionaries?
A hash table is a data structure that maps keys to values via hash functions. Dictionaries use hash tables internally for fast key-value mapping.

9. Can lists contain different data types in Python?
Yes, Python lists can hold mixed data types: [1, "two", 3.0].

10. Why are strings immutable?
Strings are immutable to ensure that stored text cannot be changed, allowing safe sharing and efficient memory usage.

11. Advantages of dictionaries over lists for certain tasks
Dictionaries allow O(1) access by key, whereas lists require O(n) searching by value. They're better for fast lookups.

12. When is a tuple preferable over a list?
When you want to protect data from changes, such as coordinates (x, y) or days of the week.

13. How do sets handle duplicate values in Python?
Sets automatically remove duplicates—each value can only appear once.

14. How does the “in” keyword work differently for lists and dictionaries?
For lists: checks if a value exists (2 in ).

For dictionaries: checks if a key exists ('name' in {'name': 'Bob'}).

15. Can you modify the elements of a tuple? Explain.
No, tuples are immutable—once created, you cannot change their content.

16. What is a nested dictionary? Example use case.
A nested dictionary is a dictionary containing another dictionary as a value.

Example: {"student": {"name": "Ana", "marks": 90}}

17. Time complexity of accessing elements in a dictionary
Average and amortized time complexity is O(1), thanks to hash tables.

18. In what situations are lists preferred over dictionaries?
When order matters or you need indexed/random access, like storing numbers or names in a certain sequence.

19. Why are dictionaries considered unordered? How does that affect data retrieval?
Dictionaries do not guarantee order before Python 3.7, so retrieval by key is fast but order is arbitrary.

20. Difference between a list and a dictionary in terms of data retrieval

 Lists use integer indices for access.

 Dictionaries use keys for retrieval, allowing direct access regardless of order.
